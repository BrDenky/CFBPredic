```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(
    r'C:\Users\ASUS\Desktop\CFBPredic\data\2.Ipc_serie_historicacanasta_basica_sep_2025\SERIE BÁSICA.csv',
    encoding='latin-1',
    header=None
)

print("CONTENIDO COMPLETO:")
print(df.to_string())
```

```python
df_limpio = df.iloc[11:433].copy()
df_limpio = pd.concat([df_limpio.iloc[:4], df_limpio.iloc[9:]])
df_limpio.columns = ['government', 'month', 'year', 'cfb_cost', 
                     'family_income', 'consumptionCI', 'RexPercent'
                     ,'col7','Col8', 'Col9', 'Col10', 'Col11']
df_limpio = df_limpio[['government', 'month', 'year', 'cfb_cost', 'family_income', 'consumptionCI', 'RexPercent']]

# Manejar nombre government en NAN
df_limpio['government'] = df_limpio['government'].ffill()


# Eliminar espacios extras al inicio y final de cada columna
columnas_texto = df_limpio.select_dtypes(include=['object']).columns

for col in columnas_texto:
    df_limpio[col] = df_limpio[col].str.strip()

# Eliminar incosistencias en meses
df_limpio['month'] = df_limpio['month'].str.replace('*', '', regex=False)

# Eliminiar porcentaje en columna y transformar a float
df_limpio['RexPercent'] = df_limpio['RexPercent'].str.replace('%', '', regex=False)
df_limpio['RexPercent'] = pd.to_numeric(df_limpio['RexPercent'], errors='coerce')
```

```python
mes_map = {
    # "January": 1, "February": 2, "March": 3, "April": 4,
    # "May": 5, "June": 6, "July": 7, "August": 8,
    # "September": 9, "October": 10, "November": 11, "December": 12,
    
    # Si tus meses están en español, usa esto en cambio:
    "Enero": 1, "Febrero": 2, "Marzo": 3, "Abril": 4,
    "Mayo": 5, "Junio": 6, "Julio": 7, "Agosto": 8,
    "Septiembre": 9, "Octubre": 10, "Noviembre": 11, "Diciembre": 12
}

# Manejo de columna month
df_limpio["month_num"] = df_limpio["month"].map(mes_map)

df_limpio["date"] = pd.to_datetime(
    df_limpio["year"].astype(str) + "-" + df_limpio["month_num"].astype(str) + "-01"
)

df_limpio = df_limpio.sort_values("date")
```
